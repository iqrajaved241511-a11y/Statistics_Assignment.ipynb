```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
marks = [78, 85, 90, 67, 88, 92, 75, 81, 95, 70]

print(marks)
```

```python
mean_value = np.mean(marks)

print("Mean:", mean_value)
```

```python
median_value = np.median(marks)

print("Median:", median_value)
```

```python
variance_value = np.var(marks)

print("Variance:", variance_value)
```

```python
std_value = np.std(marks)

print("Standard Deviation:", std_value)
```

```python
trimmed_mean = stats.trim_mean(marks, 0.1)

print("Trimmed Mean:", trimmed_mean)
```

```python
print("Minimum Marks:", np.min(marks))
print("Maximum Marks:", np.max(marks))
```

```python
print("\n--- Statistical Summary ---")

print("Mean =", mean_value)
print("Median =", median_value)
print("Variance =", variance_value)
print("Standard Deviation =", std_value)
print("Trimmed Mean =", trimmed_mean)
print("Minimum =", np.min(marks))
print("Maximum =", np.max(marks))
```
